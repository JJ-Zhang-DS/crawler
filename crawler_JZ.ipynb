```python
from urllib.request import Request, urlopen
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import numpy as np
import time
import random
import os
import json
```

```python
class Crawler():
    def get_soup(self, url):
        """
        This function get the beautifulsoup object of a webpage.

        Args:
            url (str): the link string of webpage

        Returns:
            soup (obj): beautifulsoup object
        """
        request = Request(url, headers={'User-Agent': 'Resistance is futile'})
        response = urlopen(request)
        return BeautifulSoup(response, "html.parser")
    
    def get_product_links_from_page(self, url):
        """
        This function gets the links of the jobs on the joblist page.

        Args:
            url (str): link to joblist page

        Returns:
            job_page_links (list): list of links to the webpages of the jobs
        """
             
        soup = self.get_soup(url)
        head = "http://www.walmart.com"   #check here
        links = []
        for item in soup.find_all("a", href=True):
            if 'link-identifier=' in str(item) and 'absolute' in str(item):
                link = item['href']
                links.append(head+link)
        return links
     
    def get_products(self, keyword):
        """
        Args:
            job_title (str): example: 'data+scientist'

        Returns:
        """
        job_gap_min = 5 #min sleep time between jobs
        job_gap_max = 6 #max sleep time between jobs
        
        url = "https://www.walmart.com/search?q="
        keyword_list = keyword.split()
        links = []
        for i in range(len(keyword_list)):
            url+='+'
            url+=keyword_list[i]    

            #get links to webpages of jobs on the joblist
        links = self.get_product_links_from_page(url)

        for link in links:
            gap = np.random.uniform(job_gap_min,job_gap_max) 
            time.sleep(gap)
            data = self.get_info_from_product_page(link)

            print(json.dumps(data))
            
    def get_info_from_product_page(self, url):
        """
        This function get all the useful info from the job webpage.

        Args:
            url (str): link to job webpage

        Returns:
            data (dict): dictionary with keywords: 
                         item, price, link
        """
        soup = self.get_soup(url)
        data = {}
        
        try:
            data['item'] = soup.find('h1', itemprop="name").getText()
            data["price"] = soup.find('span', itemprop="price").getText()
            data["link"] = str(url)
        except:
            pass
        return data
```
